```python
data['GPA'] = data['GPA'].replace(['Personal ', 'Unknown', '3.79 bitch'], -1).astype(float)
data['GPA'] = data['GPA'].replace(-1, data['GPA'].mean())
```

```python
data['weight'] = data['weight'].replace(["I'm not answering this. ", 'Not sure, 240','144 lbs'], -1).astype(float)
data['weight'] = data['weight'].replace(-1, data['weight'].mean())
```

```python
num_df = data.select_dtypes(include = 'number')
obj_df = data.select_dtypes(include = 'object')
obj_df
```

```python
for col in num_df.columns:
    mean = num_df[col].mean()
    num_df[col] = num_df[col].fillna(mean)

num_df
```

```python
for col in obj_df.columns:
    mode = obj_df[col].mode()[0]
    obj_df[col] = obj_df[col].fillna(mode)

obj_df
```

```python
data = pd.concat([num_df,obj_df],axis=1)
data.head()
```

```python
data.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['father_profession_encoded'] = le.fit_transform(data['father_profession'].str.lower())
data.head()
```

```python
encoding_mapping = dict(zip(data['father_profession_encoded'].unique(), data['father_profession'].str.lower().unique()))
print("Mapping between Original Values and Encoded Values:")
for original_value, encoded_value in encoding_mapping.items():
    print(f"{original_value}: {encoded_value}")
```

```python
le = LabelEncoder()
data['fav_cuisine_encoded'] = le.fit_transform(data['fav_cuisine'].str.lower())
data.head()
```

```python
encoding_mapping = dict(zip(data['fav_cuisine_encoded'].unique(), data['fav_cuisine'].str.lower().unique()))
print("Mapping between Original Values and Encoded Values:")
for original_value, encoded_value in encoding_mapping.items():
    print(f"{original_value}: {encoded_value}")
```
